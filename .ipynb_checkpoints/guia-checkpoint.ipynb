# Orígenes de la teoría del electromagnetismo

Resumido de [1]:

Desde la antigua Grecia, era conocido que ciertos materiales como el ámbar atraían a otros luego de ser frotados, y de iguall modo de las fuerzas de atracción y repulsión en trozos de magnetita, incluso se cree que la brújula fue utilizada en China mucho antes de que los griegos descubrieran las propiedades de la magnetita. En una obra publicadaen el año 1600 del médico William Gilbert se considero que la tierra podía ser considerada como un gran imán. Las primeras leyes cuantitativas sobre estos fenómenos aparecen a mediados del siglo XVIII, John Mitchell en 1750 concluyó que las fuerzas entre imanes variaban inversamente con el cuadrado de la distancia, en 1785 Charles Coulomb encontraría la misma dependencia con la distancia entre dos cargas eléctricas. En 1807 el físico Hans Oersted comienza a realizar experimentos para hallar alguna relación entre la __electricidad__ y el __magnetismo__ , en 1819 realizó un experimento en el que probó que al colocar una brújula cerca de un cable por el que circulaba __corriente__, la brújula tendía a orientarse perpendicularmente al cable, éstas ideas se difundieron hasta llegar a André-Marie Ampere presentaría una memoria en la que se indicaba que dos __conductores__ en paralelos por los que circulaba corriente eléctrica se atraían si las corrientes tenían el mismo sentido y en cambio se repelían si las corrientes iban en sentidos opuestos. En ese mismo año Jean-Baptiste Biot y Félix Savart, estudiarían el grado en el que se desviaba una aguja (como en el experimento de Oersted), encontrando la dependencia con el inverso de dicha distancia, estos experimentos demostrabanque las corrientes eléctricas podrían ser el origen del magnetismo, otros científicos se preguntaron si era posible lo opuesto: ¿podía el magnetismo generar corrientes eléctricas? Farady lo comprobaría en un famoso experimento en el que justo al abrir el interruptor de un  circuito aparecía corriente eléctrica en un circuito cercano que no estaba conectado a ninguna pila, en 1831 realizó otro experimento que dejó las cosas más claras: al mover un imán cerca de una bobina se detectaba en ella una __corriente eléctrica inducida__.  

__Algunas definiciones__

_Carga eléctrica_: Es una propiedad de la materia que tienen algunas partículas 'fundamentales' y que dictan como se afectan el campo eléctrico y el campo magnético. Hay dos tipos de cargas, positivas y negativas. La carga positiva se asocia al protón y la negativa al electrón, ambos comparten una magnitud cuantizada que equivale a $1.602176634 \times 10^{-19} C$, para los electrones es negativa y para los protones positiva. Se miden en Coulombs (C).

<center>
    <img src="img/electric_charges.png" width="250"/>
</center>

_Electricidad y magnetismo_: Son los fenómenos fenómeno producidos por la presencia o movimiento de cargas eléctricas hacen parte de un mismo fenómeno llamado electromagnetismo. Ambos fenómenos se caracterizan mediante el campo eléctrico y el campo magnético. El campo magnético se mide en Tesla (T= N/(A*m)) y el campo electrico se mide en (V/m)

<center>
    <img src="img/magnetic_field.png" width="250"/>
</center>

_Corriente eléctrica_: Es el flujo de partículas cargadas moviendose através de un conductor eléctrico o el espacio. se mide en Amperios(A). En la siguiente imágen de un circuito se representa mediante la letra $i$. 

<center>
    <img src="img/current.png" width="250"/>
</center>

_Conductor eléctrico_: Es un material que permite el flujo libre en una o varias direcciones de la corriente eléctrica. La conductancia es la propiedad que mide la habilidad de un medio para permitir el paso de la corriente y se mide en Siemens (S).

_Dieléctrico_: Es un tipo de material que permite el paso de corriente pero no con la misma libertad que un material conductor.

_Corriente eléctrica inducida_: es el fenómeno que produce una fuerza electromotriz mediante cambios o expocisión a un campo magnético variable.

_Permitividad_: Es una medida de que tanto las moléculas de un material se oponen a un campo eléctrico externo.

_Permeabilidad_: Es una medida de que tanto un medio se opone a orientación del campo magnético.

```python
import matplotlib.pyplot as plt # plotting libray
from matplotlib import cm # Color map utilities for contour plots
from mpl_toolkits.mplot3d import Axes3D # 3D plot utilities
import numpy as np # Array manipulation library
from numpy.linalg import norm # function to compute euclidean distance
#--------------------------------------------------------------------------------------THIS IS IMPORTANT, PLEASE READ ME-----------------------------------------------------------
# There are two modes to display the graphics
# [The interactive mode] 
# on this mode you can directly zoom the graph, change the perspective,and among others to use this mode you have to
# install on the anaconda environment: 
# - nodejs
# - widgetsnbextensions
# - ffmpeg
# After this installation it might be necessary to update Jupyter also in anaconda navigator comment out
#%matplotlib widget
#%matplotlib notebook
# and run the cell, up to this point any plot can be visualized as an interactive image.
# [The static mode]
# this is the default mode for most plots, if the first mode does not work this would be the only way to visualize the images,
# once again, to use it comment out the following line.
%matplotlib inline
```

```python
# The charge class describes as an object an electric charge
# it's construction requires two parameters, the location
# and the electric charge, it assumes that the position is in meters
# and the electric charge is in coulombs.
class Charge:
    def __init__(self, position, charge):
        self.position = position
        self.charge = charge

# The experiment class, is an abstract construction where several objects will interact
# and affect between each other, up to this point only an static Electric field can be computed.
class Experiment:
    # Constructor, no parameters needed
    def __init__(self):
        self.K=9e9 # This is the Coulomb constant.
        self.charges=[] # an array to store each charge.
    
    # This method adds a charge object to the experiment
    def add_charge(self,c):
        self.charges.append(c)
    
    # This method computes the electric field between a test charge and the charges that are
    # contained in the experiment on 'self.charges'. Essentially applies the vector form of
    # Coulomb's Law.
    def EField_at_charge(self, test_charge):
        e_field = 0 # the Total electric field
        v_hat=0 # The unitary vector between the charges
        # By the principle of superposition, we can sum each effect of every charge
        # in the experiment acting on the test charge.
        for c in self.charges:
            s = test_charge.position - c.position # We need the distance between the test charge and the current charge
            dist = norm(s) # Compute the euclidean distance
            v_hat = s/dist # Compute the unitary vector ()
            e_field += (experiment.K * c.charge * test_charge.charge * v_hat)/dist**2 # We apply Coulomb's law.
        return e_field, v_hat # Return the value of the electric field and the unitary vector.
    
    # This method computes the value of the electric field at somepoint given the current distribution of
    # charges in the experiment. In esscence is similar to the EField_at_charge method, with the difference
    # that it is only required the position, the method assumes that the charge is equal to 1. Additionally
    # returns the magnitude of the electric field at that point.
    def EField_at_point(self, point):
        # apply Coulomb Law
        e_field = 0
        v_hat=0
        e_field_magnitude = 0
        for c in self.charges:
            s = point - c.position
            dist = norm(s)
            v_hat = s/dist
            e_field += (experiment.K * c.charge * v_hat)/dist**2
        e_field_magnitude = norm(e_field)
        return e_field, v_hat, e_field_magnitude
    
    # This method computes the electric field at a set of points (a mesh).
    def EField_at_mesh(self, mesh):
        # this form is useful for plotting the results
        # [[x-dimension], [y-dimension], [z.dimension]]
        e_field = [[], [], []]
        v_hat = [[], [], []]
        e_field_magnitude = []
        
        #these are auxiliary variables
        ef_at = 0
        v_hat_at = 0
        ef_mag_at = 0
        
        # For each point we calculate the magnitude and direction of the E-Field with our defined function
        # 'self.EField_at_point'.
        for i in range(mesh.shape[1]):
            # Compute the results at the current point
            ef_at, v_hat_at, ef_mag_at = self.EField_at_point([mesh[0][i], mesh[1][i], mesh[2][i]])
            e_field_magnitude.append(ef_mag_at)
            
            # Structure the results in the expected (and convenient way).
            e_field[0].append(ef_at[0])
            e_field[1].append(ef_at[1])
            e_field[2].append(ef_at[2])
            
            v_hat[0].append(v_hat_at[0])
            v_hat[1].append(v_hat_at[1])
            v_hat[2].append(v_hat_at[2])            
            
        return np.array(e_field), np.array(v_hat_at), np.array(e_field_magnitude)
    
    # This utilitary function helps to plot the electric charge distribution given their positions.
    # returns the plot for further manipulation (for example,to plot the E-field, o make another useful representation).
    def plot(self, fig):
        ax = fig.add_subplot(111, projection='3d')
        
        for c in self.charges:
            color = 'b'
            if c.charge >= 0:
                color = 'r'
            ax.scatter(c.position[0],c.position[1],c.position[2], color=color)
            
        return ax
```

```python
# Test with several charge.
# The orange charges are positive charges and the blue ones are negative charges, in this example we are only plotting the value of the field at one point

# Define a new experiment
experiment = Experiment()
# Add the charges
experiment.add_charge(Charge(np.array([0,0,0]), -6)) #Charge(np.array([x-dimension,y-dimension,z-dimension]), electric charge)
experiment.add_charge(Charge(np.array([0,0,0]), -2))
experiment.add_charge(Charge(np.array([0,0,0]), -2))
experiment.add_charge(Charge(np.array([0.5,0.,0.5]), 3))

# Define a test charge
test_charge = Charge(np.array([1,1,1]), 1)

# compute the E-field at the test charge
e_field, v_hat = experiment.EField_at_charge(test_charge)

# plot results
fig = plt.figure(figsize=(10,8))
# plot the charges of the experiment
ax = experiment.plot(fig)

#Add the test charge to the plot
ax.scatter(test_charge.position[0], test_charge.position[1], test_charge.position[2], color='y')
# Add the direction of the electricfield
ax.quiver(test_charge.position[0], test_charge.position[1], test_charge.position[2], e_field[0], e_field[1], e_field[2], length=0.5, normalize=True)

# Set labels on the axis.
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-1,3)
ax.set_ylim(-1,3)
ax.set_zlim(-1,3)

plt.show()
```

```python
# Compute it as a field
# using a distribution of charges and a delimited 3-d space, compute the E-Field. It would be useful to use for this representation the
# interactive mode. if it does not work use the staticmode as defined at the begining of the file.
%matplotlib widget
%matplotlib notebook
#%matplotlib inline

experiment = Experiment()
experiment.add_charge(Charge(np.array([-1,   4.5,  0]), -6))
experiment.add_charge(Charge(np.array([ 0,   3.8, -0.1]), -2))
experiment.add_charge(Charge(np.array([ 0.5, 3.14 ,   -0.3]), -2))
experiment.add_charge(Charge(np.array([ 1,   2,    0.5]), 3))

# I used 10 to compute the graph but increasing the granularity of the mesh can cause and increase in the time to compute
# the graph interactive mode this is inadmisible, keep this in mind.
N = 10
x = np.linspace(-5,5, N)
y = np.linspace(-5,5, N)
z = np.linspace(-5,5, N)

# We define the mesh
mesh = []

xx = []
yy = []
zz = []

for z_ in z:
    for x_ in x:
        for y_ in y:
            xx.append(x_)
            yy.append(y_)
            zz.append(z_)

mesh = np.array([xx, yy, zz])

# Compute the E-Field at the mesh
e_field, v_hat, e_mag = experiment.EField_at_mesh(mesh)
#plot the results
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
# Plot the E-field on the mesh.
ax.quiver(mesh[0],mesh[1], mesh[2], e_field[0], e_field[1], e_field[2], length=0.5, normalize=True)
plt.show()
```

# Las ecuaciones de Maxwell

$$\vec{\nabla} \cdot \vec{D} = \rho,$$
$$\vec{\nabla} \cdot \vec{B} = 0,$$
$$\vec{\nabla} \times \vec{E} = - \frac{\partial \vec{B}}{\partial t},$$
$$\vec{\nabla} \times \vec{H} = \vec{J} + \frac{\partial \vec{D}}{\partial t}.$$

Las ecuaciones como vemos, se describen en términos de dos operadores matemáticos, la divergencia y el rotacional. La divergencia procede del concepto de flujo, de hecho construyó el concepto formal imaginando que aquellas perturbaciones que creaban las cargas y corriente en cada punto del espacio, debía comportarse como un fluido incompresible. Maxwell pensó  que las lineas de fuerza de una carga eléctrica debían comportarse similar a un fluido, de esta manera considero que las cargas positivas pueden ser fuentes que 'emanan' campo en todas las direcciones y por igual

<center>
    <img src="img/gauss_law.png" width="250"/>
</center>

Si imaginamos una esféra con centro en la carga y con radio $r$, las lineas de campo serán perpendicularesa la esfera el flujo del campo se puede calcular mediante

$$\Phi_e = 4\pi r^2E,$$

Despejando $E$ tenemos el campo elécrico de una carga puntual, y utilizando la ley de Coulomb $E = Q/4\pi\epsilon_0r^2$ igualando obtenemos que

$$\Phi_e = \frac{Q}{\epsilon_0}$$.

Por otro lado elexperimento de Oersted demostró que una corriente eléctrica crea en su entorno un campo magnético, adicional a ello experimentalmente hay una diferencia entre las lineas de fuerza generadas por cargas eléctricas y las lineas de fuerza magnéticas creadas por un cable de corriente, las primeras emanan, las segundas parecen cerradasy en muchos casos no hay un punto en donde comiencen y terminen, por lo tanto el flujo de estas líneas debe ser nulo [1].

Mediante esta operación si imaginamos un cubo a través del cual está fluyendo un campo eléctrico, las componentes del campo que describe el desplazamiento eléctrico $\vec{D}$ tendría componentes $D_x, D_y, D_z$ y su divergencia sería

$$\frac{\partial D_x}{\partial x} + \frac{\partial D_y}{\partial y} + \frac{\partial D_z}{\partial z},$$

Y deacuerdo con el teorema de Gauss, el flujo del vector de desplazamiento eléctrico coincide con la carga dentro de un volúmen imaginario, al dividir esta carga por el volúmen en el límite se reduce a cero por lo que llegamos a la igualdad descrita en la ley de Gauss para campos electrostáticos. Repitiendo el mismo procedimiento en laslineas de fuerza del campo magnético al ser curvas cerradas, la diferencia en el límite siempre será 0 y es lo descrito en la Ley de Gauss para campos magnéticos.

# Ley de Gauss para campos electrostáticos

_El campo eléctrico producido por una carga eléctrica diverge en las cargas positivas y converge sobre las cargas negativas_.

$$\vec{\nabla} \cdot \vec{D} = \rho,$$

Vamos a descomponer la ecuación, lo principal que es una ecuación diferencial en forma vectorial, $\nabla$ es el operador de divergencia, la divergencia nos permite cuantificar la tendencia a divergir de este campo eléctrico, $\vec{D}$ que es el vector de desplazamiento eléctrico o la densidad eléctrica del flujo y se relaciona con el campo eléctrico como 

$$\vec{D} = \epsilon\vec{E},$$

donde $\epsilon$ es la permitividad eléctrica del material o constante dieléctrica. y $\vec{E}$ el campo eléctrico. $\rho$ es la densidad de carga eléctrica describe la cantidad de carga eléctrica en un volúmen. 

# Ley de Gauss para campos magnéticos.

_La divergencia del campo magnético en cualquier punto es cero_.

$$\vec{\nabla} \cdot \vec{B} = 0,$$

Nuevamente descomponemos la expresión, en este caso tenemos $\vec{B}$, éste es el campo magnético, su escritura nos dice que las cargas magnéticas no existen.

Faraday probó que la condición necesaria para que hubiera corriente inducida en un circuito era que el flujo del campo magnético que lo atravesaba cambiara en el tiempo, por lo tanto en ese punto $\partial \vec{B}/ \partial t \neq 0$, si colocamos una espira diminuta de material conductor en esa región del espacio aparecerá una corriente inducida que está generada por un campo eléctrico.

# Ley de Faraday

_Un campo eléctrico circulante es producido por un campo magnético que cambia con el tiempo_.

$$\vec{\nabla} \times \vec{E} = - \frac{\partial \vec{B}}{\partial t},$$

El rotacional de un campo vectorial es una medida de cual es la tendencia de un campo a circular alrededor de un punto en este caso el del campo eléctrico, al otro lado de la igualdad tenemos que el cambio del cambo magnetico a traves del tiempo es proporcional pero en el sentido contrario.

Volviendo a la analogía de un fluído si el campo se puede volver un fluido, como en la imágen

<center>
    <img src="img/ampere.png" width="250"/>
</center>

podemos decir que en el punto por donde pasa la corriente (y que en principio es diferente de cero) el rotacional del campo magnético también será diferente de cero, pero además el rotacional será perpendicular a las lineas de campo y a la corriente y por tanto son paralelos, originalmente la ley de Ampere establece que la constante de proporcionalidad entre ambos es la permeabilidad magnética del medio($\mu$) y en forma diferencial se describe como

$$\vec{\nabla} \times \vec{B} = \mu\vec{J}$$

finalmente Maxwell deduce que es necesario incluir un término adicional al que llamó corriente de desplazamiento, para tener en cuenta los pequeños desplazamientos que podía tener cargas eléctricas dejando así finalmente la ley de Ampere-Maxwell.

# Ley de Ampere-Maxwell

_Un campo magnético circulante es producido por una corriente y por un campo eléctrico que cambia a través del tiempo_.

$$\vec{\nabla} \times \vec{H} = \vec{J} + \frac{\partial \vec{D}}{\partial t}.$$

Donde $\vec{H}$ es la densidad de flujo magnético y se describe como
$$\vec{H} = \frac{\vec{B}}{\mu}.$$
